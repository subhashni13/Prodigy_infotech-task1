```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python

# Load the dataset
data = pd.read_csv('data.csv')
data
```

```python

# Remove duplicate rows
cleaned_data = data.drop_duplicates()

# Save the cleaned dataset to a new file (optional)
cleaned_data.to_csv('cleaned_data.csv', index=False)

# Display the cleaned data
print(cleaned_data)
```

```python
# Iterate over each column to generate bar charts
for column in data.columns:
    if data[column].dtype in ['int64', 'float64']:  # Check if the column is numerical
        plt.figure(figsize=(10, 6))
        plt.bar(data.index, data[column], color='skyblue')
        plt.xlabel('Index')
        plt.ylabel(column)
        plt.title(f'Bar Chart for {column}')
        plt.show()
```

```python
# Iterate over each column to generate histograms
for column in data.columns:
    if data[column].dtype in ['int64', 'float64']:  # Check if the column is numerical
        plt.figure(figsize=(10, 6))
        plt.hist(data[column].dropna(), bins=30, color='skyblue', edgecolor='black')
        plt.xlabel(column)
        plt.ylabel('Frequency')
        plt.title(f'Histogram for {column}')
        plt.show()
```
